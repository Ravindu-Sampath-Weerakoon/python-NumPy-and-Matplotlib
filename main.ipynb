```python
import numpy as np 
import matplotlib.pylab as plt 
```

```python
# Creating a new array, filled with zeros.
n= 10 
a_zeros = np.zeros(n)
print(a_zeros)
a_zeros_int = np.zeros((n,), dtype=int)
print(a_zeros_int)
```

```python
# To generate a sequence of Random Numbers
# np.random.normal(mean,sd,size)
array_normal = np.random.normal(5,1,100000000)
# print(array_normal)
print(np.average(array_normal))
```

```python
arrayStatistical = np.random.normal(10,20,100)
print("min = " ,np.min(arrayStatistical))
print("max= " ,np.max(arrayStatistical))
print("mean = " ,np.mean(arrayStatistical))
print("median = " ,np.median(arrayStatistical))
print("standard deviation = " ,np.std(arrayStatistical))
```

```python
# Product of two NumPy arrays
a = np.array([[1,2],[3,4]])
b = np.array([[1,2],[3,4]])
```

```python
print(a*b)
```

```python
print(np.dot(a,b)) 
print(np.matmul(a,b))
print(np.dot(a,b) == np.matmul(a,b))
```

```python
# matplotlib.pyplot
plt.plot([1,2,3,4],[1,2,4,9])
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
# Sub-Plots

x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.plot(x,y)
plt.title('1st subplot')
plt.xlabel('x axis')    
plt.ylabel('y axis')
plt.subplot(1,3,2)
plt.plot(x,y)
plt.title('2nd subplot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.subplot(1,3,3)
plt.plot(x,y)
plt.title('3rd subplot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
# Plots Features

x1 = [1, 2, 3, 4]
y1 = [1, 2, 4, 9]
x2 = [1, 2, 3, 4]
y2 = [1, 4, 9, 16]

# Create the plot
plt.plot(x1, y1, label='Line 1')  # Plot for the first line with a label
plt.plot(x2, y2, label='Line 2', linestyle='--')  # Plot for the second line with a different style and label

# Customize the plot
plt.title("This is a title" , fontsize=16, fontweight='bold' , color ='blue')
plt.xlabel("X axis")
plt.ylabel("Y axis")

# Add a legend to differentiate between lines
plt.legend()

# Show grid for better readability
plt.grid(True)

# Display the plot
plt.savefig('plot.png')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ko-')  
plt.show()
```

```python
#Bar Plot
x=np.array([1,2,3,4])
y=np.array([5,10,4,1])
plt.bar(x,y , color ='k')
plt.show()
```

```python
# Histogram & Scatterplot
plt.hist([1,2,4,5,5])
plt.show()
```

```python
# Sample data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

# Plot histogram with lines connecting midpoints of bins
plt.hist(data, bins=2, histtype='step', label='Data')

# Add labels and title for clarity
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram using Line Style')
plt.legend()  # Show legend to identify the data source

# Display plot
plt.show()
```
